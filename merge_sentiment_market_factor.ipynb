```python
merged_factors = pd.merge(market_factors,sentiment_factors,how = 'left',left_on = ['datetime','assetCode'],right_on = ['datetime','assetCode'])
```

```python
merged_factors.columns
```

```python
merged_factors['alert'].fillna(0,inplace=True)
merged_factors['article'].fillna(0,inplace=True)
```

```python
merged_factors['news_coverage_flag'] = merged_factors['alert']+merged_factors['article']*3
```

```python
merged_factors['news_coverage_flag']
```

```python
merged_factors_fillna = copy.deepcopy(merged_factors)
for signal in list(sentiment_factors.columns)[6:-1]:
    merged_factors_fillna[signal].fillna(merged_factors_fillna[signal].median())
```

```python
with open('merged_factors.pkl','wb') as f:
    pickle.dump(merged_factors, f, pickle.HIGHEST_PROTOCOL)
```
